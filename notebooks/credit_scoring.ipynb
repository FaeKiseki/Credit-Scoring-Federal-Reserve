```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import streamlit as st

# Display configuration
pd.set_option('display.max_columns', 100)

# Load dataset
df = pd.read_csv("../data/24Q4-CreditCardBalances.csv")

# Show first 5 rows
df.head()
```

```python
# General informations

# Dataset shape
print("Shape:", df.shape)

# Column types and missing values info
df.info()

# Descriptive statistics
df.describe(include='all')
```

```python
# Percentage of missing values per column
missing = df.isnull().mean().sort_values(ascending=False) * 100
missing[missing > 0]
```

```python
# Visualize distributions of key numerical variables with rotated and smaller labels

# Matplotlib style parameters
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10

# Columns to visualize
cols = [
    'Account Balance (Active Accounts Only) (50th percentile)',
    'Account Balance (Active Accounts Only) (75th percentile)',
    'Current Credit Limit (50th Percentile)',
    'Current Credit Limit (75th Percentile)',
    'Current Credit Score (50th percentile)',
    'Utilization (Active Accounts Only) (50th percentile)',
    'Utilization (Active Accounts Only) (75th percentile)'
]

# Plot loop
for col in cols:
    plt.figure()
    sns.histplot(df[col].dropna(), kde=True, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of: {col}', pad=20)
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.xticks(rotation=90, ha='right')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
```

```python
# Correlation matrix

# Selecting only numeric columns
numeric_df = df.select_dtypes(include=np.number)

# Computing the correlation matrix
corr_matrix = numeric_df.corr()

# Plotting the heatmap
plt.figure(figsize=(20, 16))
sns.heatmap(
    corr_matrix,
    cmap='coolwarm',
    annot=True,
    fmt=".2f",
    linewidths=0.5,
    square=True,
    cbar_kws={'shrink': 0.8}
)

# Plot aesthetics
plt.title("Correlation Matrix of Numerical Features", fontsize=18, pad=20)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
def clean_numeric_value(value):
    """
    Clean values containing symbols like $, %, or commas,
    remove extra spaces, and convert to float.
    """
    if isinstance(value, str):
        return float(value.replace('$', '').replace('%', '').replace(',', '').strip())
    return value

# Apply the cleaning function to all columns except 'YRQTR'
for col in df.columns:
    if col != 'YRQTR':
        df[col] = df[col].apply(clean_numeric_value)

# Verification after cleaning
print("\nVerification after cleaning:")
print(df[['Total Balances ($Billions)', 'Utilization (Active Accounts Only) (50th percentile)']].head())

# Optional: check data types
print("\nData types:")
print(df.dtypes[['Total Balances ($Billions)', 'Utilization (Active Accounts Only) (50th percentile)']])
```

```python
# Convertion of 'YRQTR' into usable date features

import re

# Filter rows with valid 'YRQTR' format YYYYQ1-4
valid_format = df['YRQTR'].apply(lambda x: isinstance(x, str) and re.fullmatch(r'\d{4}Q[1-4]', x) is not None)
df = df[valid_format].copy()

# Create PeriodIndex from 'YRQTR' strings
period_index = pd.PeriodIndex(df['YRQTR'], freq='Q')

# Convert PeriodIndex to timestamps (start of quarter)
df['date'] = period_index.to_timestamp()

# Set 'date' as index
df.set_index('date', inplace=True)

# Sort dataframe by date
df.sort_index(inplace=True)

# Return first 5 rows to verify
df.head()
```

```python
# Display basic info about the dataframe
print(df.info())

# Show first 5 rows to verify content
print(df.head())

# Check for missing values per column
print("Missing values per column:")
print(df.isnull().sum())

# Summary statistics of numerical columns
print("Summary statistics:")
print(df.describe())
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['Total Balances ($Billions)'], color='royalblue', linewidth=2)
plt.title("Total Credit Card Balances Over Time", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Balances ($ Billions)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
